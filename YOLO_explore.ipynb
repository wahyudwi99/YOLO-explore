```python
# Drawing bounding box manually
from numpy import asarray
%matplotlib inline

load_image = Image.open('/content/drive/MyDrive/Pendidikan/Data Science/Project/Tensorflow_Projects/Computer Vision/YOLO_NEW_1/7104-1.jpeg')
image_array = asarray(load_image)

image = cv2.rectangle(image_array, (65,78), (502,551), (255,0,0), 2)
cv2.putText(image, 'Sunflower ' + '63%', (65, 78-6), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (40,200,10), 3, cv2.LINE_AA)
plt.imshow(image)
plt.show()
```

```python
image_array.shape
```

### **Many objects**

```python
# Custom object detection

%matplotlib inline

# gambar = Image.open('/content/49830969241_b6c7249693_o.jpeg')
gambar = Image.open('/content/drive/MyDrive/Pendidikan/Data Science/Project/Tensorflow_Projects/Computer Vision/YOLO_NEW_1/download.jpeg')
gambar_array = asarray(gambar)

prediction_1 = model.predict(gambar_array)

kotak = prediction_1[0].boxes
kotak_nilai = kotak.xyxy

kotak_list = kotak_nilai.tolist()

for a in range(len(kotak_nilai)):
  # if kotak.boxes.tolist()[a][5] != 0:
  # xmin, ymin = kotak_list[i][0], kotak_list[i][1]
  # xmax, ymax = kotak_list[i][2], kotak_list[i][3]
  if kotak.cls[a] != 0:
    color = 255,0,0

    new_kotak_1 = []
    for i in range(len(kotak_nilai.tolist())):
      temporal = []
      for j in range(len(kotak_nilai.tolist()[i])):
        number = int(kotak_nilai.tolist()[i][j])
        temporal.append(number)
    
      new_kotak_1.append(temporal)

    gambar_hasil = cv2.rectangle(gambar_array, (new_kotak_1[a][0], new_kotak_1[a][1]), (new_kotak_1[a][2], new_kotak_1[a][3]), color, 2)

    conf = []
    obj = []
    for z in range(len(kotak.boxes.tolist())):
      conf.append(kotak.boxes.tolist()[z][4])
      obj.append(kotak.boxes.tolist()[z][5])


    cv2.putText(gambar_hasil, str(prediction_1[0].names[obj[a]]) + ' ' + str(np.round(conf[a]*100, 1)) + '%', (new_kotak_1[a][0], new_kotak_1[a][1]-6), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,255,0), 1, cv2.LINE_AA)

plt.imshow(gambar_hasil)
plt.show()
```

```python
import numpy as np
from numpy import asarray
import tensorflow as tf
from tensorflow import keras
from PIL import Image
from ultralytics import YOLO

image = tf.keras.utils.load_img('/content/49830969241_b6c7249693_o.jpeg')
image_array = asarray(image)

model = YOLO('yolov8s.pt')
```

```python
prediction = model.predict(image_array)
```

```python
image_show = Image.fromarray(prediction[0].plot())
image_show.show()
```

```python
object_detected = []
object_unique = []

for i in range(len(prediction[0].boxes.data.tolist())):
  object_id = prediction[0].boxes.data.tolist()[i][5]

  object_detected.append(prediction[0].names[object_id])

  if prediction[0].names[object_id] not in object_unique:
    object_unique.append(prediction[0].names[object_id])

# for i in object_detected:
#   if object_detected(i) 
```

```python
# Draw bounding box manually

image_2 = Image.open('/content/49830969241_b6c7249693_o.jpeg')
image_2 = asarray(image_2)
image_2_resize = cv2.resize(image_2, (640, 640))

cv2.rectangle(image_2_resize, (49, 152), (210, 533), (255,0,0), 2)
plt.imshow(image_2_resize)
plt.show()
```

```python
# Inverse bounding boxes scale

image_2 = Image.open('/content/49830969241_b6c7249693_o.jpeg')
image_2 = asarray(image_2)

y_ratio = image_2.shape[0]/image_2_resize.shape[0]
x_ratio = image_2.shape[1]/image_2_resize.shape[1]

cv2.rectangle(image_2, (int(49*x_ratio), int(152*y_ratio)), (int(210*x_ratio), int(533*y_ratio)), (255,0,0), 2)
plt.imshow(image_2)
plt.show()
```
